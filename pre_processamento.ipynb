# Pré-Processamento dos dados

## Extração de informações

```python
# Criando a coluna titulação para conseguir categorizar melhor os indivíduos
# Lembrando que o padrão para nome no dataset está como: Sobrenome, Título. Nome Composto
df_train['title_name'] = df_train['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())

df_test['title_name'] = df_test['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())

# O unique vai retornar os valores únicos para a titulação
df_train['title_name'].unique()
```

```python
# Checagem no dataframe de teste também
df_test['title_name'].unique()
```

```python
def categorize_titles(title):
    if title in ['Mr', 'Mrs', 'Miss', 'Master']:
        return title
    else:
        return 'Other'

# Função para padronizar as titulações. Serão utilizadas somente 'Mr', 'Mrs', 'Miss', 'Master', o restante será classificado como 'Other'
df_train['title_name'] = df_train['title_name'].apply(categorize_titles)

df_test['title_name'] = df_test['title_name'].apply(categorize_titles)
```

```python
# Será removida a coluna 'Nome', pois ela já não é mais útil para o modelo
df_train = df_train.drop('Name', axis=1)

df_test = df_test.drop('Name', axis=1)
```

```python
# Será criada a coluna tamanho da família, através do número de filhos/pais (SibSp), e cônjuges (Parch)
df_train['family_size'] = df_train['SibSp'] + df_train['Parch']

df_test['family_size'] = df_test['SibSp'] + df_test['Parch']

# Checagem de valores únicos
df_train['family_size'].unique()
```

## Tratando os dados faltantes

```python
# Verificação de valores faltantes
df_train.isna().sum() / df_train.shape[0]*100
```

```python
# Ordena o DataFrame baseado na coluna 'title_name'
df_train.sort_values(by='title_name', inplace=True)

# Extrai a coluna 'Age' como um array 2D para realizar a imputação
# imputação é um termo que denota um procedimento que substitui os valores desconhecidos de um conjunto de dados por valores plausíveis
age_data = df_train['Age'].values.reshape(-1, 1)

# Inicializando o KNN imputer com k=5 (número de vizinhos próximos)
# KNN é um modelo que será visto nas próximas aulas
imputer = KNNImputer(n_neighbors=5)

# Realiza a imputação na coluna Age
df_train['Age'] = imputer.fit_transform(age_data)
```

```python
# Removendo a coluna 'Cabin' pois ela tem 77% de dados faltantes.
df_train = df_train.drop('Cabin', axis=1)
```

```python
# Inicializando o SimpleImputer com a estratégia de valor mais frequente
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

# Alterando a estrtura da coluna (Reshape) para a imputação (necessários para arrays com uma dimensão)
imputed_column = imputer.fit_transform(df_train['Embarked'].values.reshape(-1, 1))

# Achatamento (Flatten) a coluna de duas dimensões imputada para uma dimensão
imputed_column = imputed_column.flatten()

# Trocando os dados da coluna original para os novos valores
df_train['Embarked'] = imputed_column
```
